### operators

3. relational operators---->,<,=,<=,>=,==,!=

```python
n1 = 12
n2 = 21
print(n1+n2)
print(n1-n2)
print(n1*n2)
print(n1/n2)
print(n1**n1)
print(n1//n2)
print(n1%n2)
```

```python
n = int(input("enter the number"))
if n%3 == 0 and n%5 == 0:
    print("multiple of 3 and 5")
else:
    print("none")
```

## check if the given numbers are +ve or -ve or 0

```python
n = int(input("enter the number"))
if n>0:
    print("positive")
elif n<0:
    print("negative")
else:
    print("0")
    
```

## leap year or not

```python
year = int(input("enter the year"))
if year % 400 == 0 or (year %100 != 0 and year % 4 == 0):
    print("leap year")
else:
    print("non leap year")
```

## loops
### 1.while
### 2.for

```python
# print no from 1 to n
n = int(input("enter the number"))
i=1
while i <= n:
    print(i,end="-")
    i=i+1
```

```python
# output should be sum of even numbers from 1 to n
n = int(input("enter the number:"))
i = 1
s = 0
while i <= n:
    if i%2 == 0:
        s = s+i
    i = i+1
print(s)
    
```
